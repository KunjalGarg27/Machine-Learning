# Handling null values

```python
import numpy as np
import pandas as pd
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df=pd.read_csv('car_dataset.data',names=headers)
print(df)
```

```python
#replace ? to Nan
df.replace("?",np.nan, inplace=True)
print(df.head(5))
```

```python
missing_data=df.isnull()
print(missing_data.head(5))
```

```python
# value_counts() gives the count of each unique value un the column
for column in headers:
  print(column)
  print(missing_data[column].value_counts())
  print("")
  # based on tyhe no of missing values we can find out what do we need to do inorder replace the missing values(like mean, drop the data (whole column or row) ,etc)
  # False means there is not null value , true means null value
```

```python
# average of normalized -losses column
avg_norm_loss=df["normalized-losses"].astype('float').mean(axis=0)
print("Average of Normalized losses: ",avg_norm_loss)
print("")
df["normalized-losses"].replace(np.nan,avg_norm_loss,inplace=True)
```

```python
avg_bore=df["bore"].astype(float).mean(axis=0)
print("Average of bore: ",avg_bore)
print("")
df["bore"].replace(np.nan,avg_bore,inplace=True)
```

```python
avg_stroke=df["stroke"].astype('float').mean(axis=0)
print("Average of stroke: ",avg_bore)
print("")
df["stroke"].replace(np.nan,avg_stroke,inplace=True)
```

```python
avg_horsepower=df["horsepower"].astype('float').mean(axis=0)
print("Average of horsepower: ",avg_bore)
print("")
df["horsepower"].replace(np.nan,avg_horsepower,inplace=True)
```

```python
avg_peak_rpm=df["peak-rpm"].astype('float').mean(axis=0)
print("Average of peak-rpm: ",avg_bore)
print("")
df["peak-rpm"].replace(np.nan,avg_peak_rpm,inplace=True)
```

```python
# in num-of-doors column we have float values and 2 unique values are there two and four . we are by our choice replacing the null values with four
print(df["num-of-doors"].value_counts())
df["num-of-doors"].replace(np.nan,"four",inplace=True)
```

```python
# we are dropping all the rows which have price data
# we are removing this as price is the target variable (means this needs to be predicted with the help of the other predictors)
df.dropna(subset=["price"],axis=0,inplace=True)
#reset index , because we have deleted 4 rows
df.reset_index(drop=True, inplace=True)
print(df)
```
